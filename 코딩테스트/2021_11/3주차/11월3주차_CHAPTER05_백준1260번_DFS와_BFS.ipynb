```python
from collections import deque
import sys

#sys 사용하여 입력받기
read = sys.stdin.readline

N,M,V = sys.stdin.readline

graph = [[0]*(N+1) for _ in range(N+1)]
for _ in range(M):
  a, b = map(int, read().split())
  graph[a][b] = graph[b][a] = 1

#DFS
def dfs(v):
  visited[v]=True
  print(v, end=' ')
  for i in range(1, N+1):
    if not visited[i] and graph[v][i] == 1:
      dfs(i)

#BFS
#qeue로 구현
def bfs(v):
  queue = deque([v])
  visited[v] = True

  while queue:
    v = queue.popleft()
    print(v, end=' ')

    for i in range(1,N+1):
      if not visited[i] and graph[v][i]==1:
        queue.append(i)
        visited[i] = True

visited = [False] * (N+1)
dfs(V)
print()
visited = [False] * (N=1)
bfs(V)
```
